```python
from transformers import AutoProcessor, BarkModel
import scipy

processor=AutoProcessor.from_pretrained("suno/bark")
model=BarkModel.from_pretrained("suno/bark")
model.to("cuda")


def generate_audio(text, preset, output):
  inputs= processor(text,voice_preset=preset)
  for k, v in inputs.items():
    inputs[k]=v.to("cuda")
  audio_array=model.generate(**inputs)
  audio_array=audio_array.cpu().numpy().squeeze()
  sample_rate=model.generation_config.sample_rate
  scipy.io.wavfile.write(output,rate=sample_rate,data=audio_array)

generate_audio(text="hi, you are speaking to your clone", preset="v2/en_speaker_9", output="output.wav")


```

```python
text_prompt="""♪ hi, you are speaking to your clone ♪"""
generate_audio(text=text_prompt, preset="v2/en_speaker_9", output="music.wav")
```

```python
from TTS.tts.configs.bark_config import BarkConfig
from TTS.tts.models.bark import Bark
from scipy.io.wavfile import write as write_wav

config = BarkConfig()
model = Bark.init_from_config(config)
model.load_checkpoint(config, checkpoint_dir="bark/", eval=True)
model.to("cuda")

text_prompt="hi, how are you"
output_dict = model.synthesize(text_prompt, config, speaker_id="random", voice_dirs="bark_voices",temperature=0.95)


write_wav("output.wav",24000,output_dict["wav"])
```
